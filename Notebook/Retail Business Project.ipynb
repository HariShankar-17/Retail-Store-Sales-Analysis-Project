### 🔹 PHASE 1: Understanding Overall Business Performance

**(Warm-up, confidence building)**

Q1. How is our revenue trending over time?

Are we growing, declining, or stable month by month?

Identify peak months and low-performing months.

📊 Expected Plot

plt.plot() (time series)

🧠 What I’m checking

Can you group data by month/year?

Can you show trends clearly?


```python
copied_retail_store_datas['Order_Date'] = copied_retail_store_datas['Order_Date'].astype('datetime64[ns]')
copied_retail_store_datas['Order_Date'] 
```

```python
copied_retail_store_datas['Month_Year'] = copied_retail_store_datas['Order_Date'].dt.to_period('M')
copied_retail_store_datas['Month_Year']
```

```python
month_wise_revenue = copied_retail_store_datas.groupby('Month_Year')['Revenue'].sum()
month_wise_revenue
```

```python
plt.plot(month_wise_revenue.index.astype(str),
         month_wise_revenue.values)
plt.grid(True)
plt.xticks(rotation=85)
plt.show()
```

#### TypeError: float() argument must be a string or a real number, not 'Period'
**This error means Matplotlib doesn’t understand Period values.**


Q2. Which product categories generate the highest total revenue?

Rank categories from highest to lowest.

Identify top 1 and bottom 1 category.

📊 Expected Plot

plt.bar()

🧠 CEO Insight I Expect

“Electronics dominates revenue” (hint: pricing effect)

```python
product_ranking = copied_retail_store_datas.groupby(by='Product')['Revenue'].sum().sort_values(ascending=False)
product_ranking
```

```python
plt.bar(product_ranking.index,product_ranking.values)
plt.title("Revenue by Category")
plt.ylabel("Revenue")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

Q6. How does discount percentage affect revenue?

Do higher discounts always increase revenue?

Or do we lose money after a point?

📊 Expected Plot

plt.scatter()

🧠 CEO Insight

This tells me whether discounts are smart or risky.


```python
discount_revenue = (
    copied_retail_store_datas
    .groupby('Discount_Percent')['Revenue']
    .sum()
    .reset_index()
    .sort_values('Discount_Percent')
)
discount_revenue
```

```python
plt.scatter(copied_retail_store_datas['Discount_Percent'], copied_retail_store_datas['Revenue'])
plt.xlabel("Discount %")
plt.ylabel("Revenue")
plt.title("Discount vs Revenue")
plt.show()
```

**Insights : 5% discount made 12.15M ⬆️ (highest)**

**🔍 Insight**

**Revenue increases from 0% → 5%**

**Revenue drops after 5%**

**Higher discounts do NOT increase revenue**

**🧠 How to explain this (short & sweet)**

**“A small discount (5%) increases revenue by boosting sales volume.**

**But higher discounts reduce revenue, meaning we lose money after a certain point.”**
